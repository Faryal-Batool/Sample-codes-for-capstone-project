```python
import matplotlib.pyplot as plt
import librosa, librosa.display
import IPython.display as ipd
import numpy as np
```

```python
Spark_paper='/content/Spark_Train_1.wav'
spark_paper,sr=librosa.load(Spark_paper,sr=32000)
```

```python
# for power spectrum
def plot_magnitude_spectrum(signal, sr, title, f_ratio=1):
  # Short FFT, no overlap, nfft=8192, Hanning window
  S = librosa.stft(y=signal,n_fft=1024,window='hann')
  # Power in dB
  X_mag = librosa.power_to_db(np.abs(S)**2)
    #X = np.fft.fft(signal)
    #X_mag = np.absolute(X)

  plt.figure(figsize=(18, 5))

  f = np.linspace(0, sr, len(X_mag))
  f_bins = int(len(X_mag)*f_ratio)

  plt.plot(f[:f_bins], X_mag[:f_bins])
  plt.xlabel('Frequency (Hz)')
  plt.ylabel('Power')
  plt.title(title)
```

```python
# for log amplitde spectrum
def plot_magnitud_spectrum(signal, sr, title, f_ratio=1):
    X = np.fft.fft(signal)
    X_mag = librosa.power_to_db(np.absolute(X))

    plt.figure(figsize=(18, 5))

    f = np.linspace(0, sr, len(X_mag))
    f_bins = int(len(X_mag)*f_ratio)

    plt.plot(f[:f_bins], X_mag[:f_bins])
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude (db)')
    plt.title(title)
```

```python
plot_magnitude_spectrum(spark_paper, sr, 'Spark Drone', 0.1)
```
